```python
import yfinance as yf
```

```python
#google ın verilerine erişmek için
#akbank için AKBNK.IS yazılır, tesla için TSLA
#bir günlük periodla yaklaşık 1 yıllık veriyi çekiyoruz

google = yf.Ticker("GOOGL")

df = google.history(
    start="2025-01-01",
    end="2026-01-01",
    interval="1d"
)
```

```python
df.tail()
```

```python
df['Close'].plot() #kapanış fiyatlarının grafiği
#google ın 14 aylık borsa fiyatlarının gidişatı
```
